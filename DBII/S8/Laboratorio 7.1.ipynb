```python
import numpy as np
```

## Scores

```python
def dotProduct(Q, Doc):
    qi = np.log10(1 + Q)
    di = np.log10(1 + Doc)

    result = np.dot(qi, di)
    print(result)
    return result

def cosine(Q, Doc):
    nqi = Q / np.linalg.norm(Q)
    ndi = Doc / np.linalg.norm(Doc)

    result = np.dot(nqi, ndi)
    print(result)
    return result
```

## P1. Ejercicio 1

```python
def retrieval1(Query, Collection, func_score):
    if(func_score == dotProduct):
        for Doc in Collection:
            dotProduct(Query, Doc)
    elif(func_score == cosine):
        for Doc in Collection:
            cosine(Query, Doc)
    else: 
        print("Imma teapot")
    
collection = [
    np.array([15, 5, 20, 25]),    
    np.array([30, 0, 22, 0]),
]
query = np.array([115, 10, 2, 0])    
```

```python
retrieval1(query, collection, dotProduct)
```

```python
retrieval1(query, collection, cosine)
```
